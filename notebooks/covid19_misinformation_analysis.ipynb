# **Loading dataset: covid_fake_news**

Source: https://huggingface.co/datasets/nanyy1025/covid_fake_news

*insert information*

```python
# Loading dataset covid_fake_news
dataset1 = load_dataset("nanyy1025/covid_fake_news")
print(dataset1)
```

# **Loading dataset: covid_fact_checked_polifact**

Source: https://huggingface.co/datasets/justinqbui/covid_fact_checked_polifact

*insert information*

```python
# Loading dataset covid_fact_checked_polifact
dataset2 = load_dataset("justinqbui/covid_fact_checked_polifact")
print(dataset2)
```

# **Model Evaluation**

Dataset Creation

```python
from datasets import load_metric
from datasets import list_metrics

# store datasets & reference answers in variables
ref_vars1 = []
dataset1_list = []

ref_vars2 = []
dataset2_list = []

# dataset 1: covid_fake_news, uses "real" and "fake" labels
dataset1_data = dataset1['test'].shuffle().select(range(1000))

for i in range(len(dataset1_data)): # each tweet entry is a dict of format {id, tweet, label}
  if dataset1_data[i]['label'] == "real":
    # 0 = NOT misinformation
    ref_vars1.append(0)
    dataset1_list.append([dataset1_data[i]['tweet'], 0])
  else:
    # 1 = misinformation
    ref_vars1.append(1)
    dataset1_list.append([dataset1_data[i]['tweet'], 1])

# dataset 2: covid_fact_checked_polifact, uses Politifact scale
dataset2_data = dataset2['train'].shuffle().select(range(1000))
d2_ratings = ['true', 'mostly-true', 'half-true', 'barely-true', 'false', 'pants-fire', 'full-flop']

for i in range(len(dataset2_data)): # each tweet entry is a dict of format {row #, claim, rating, adjusted rating}
  if dataset2_data[i]['adjusted rating'] == "true":
    # 0 = NOT misinformation
    ref_vars2.append(0)
    dataset2_list.append([dataset2_data[i]['claim'], 0, dataset2_data[i]['rating']])
  else: # if marked as "misleading" or "false," we will consider these to be MISINFORMATION.
    # 1 = misinformation
    ref_vars2.append(1)
    dataset2_list.append([dataset2_data[i]['claim'], 1, dataset2_data[i]['rating']])
print(dataset2_list[0])
```

Model 1: 'spencer-gable-cook/COVID-19_Misinformation_Detector'

Source: https://huggingface.co/spencer-gable-cook/COVID-19_Misinformation_Detector

*insert information*

```python
# Model 1 on Dataset 1

sentiment_classifier = pipeline(model='spencer-gable-cook/COVID-19_Misinformation_Detector', device = 'cpu')



data1 = []

# compute predictions and store in a variable
for i in range(len(dataset1_list)):
  data1.append(dataset1_list[i][0])
predictions = sentiment_classifier(data1) # predictions is a list of dicts of format {label, score}
print(predictions[0])

predicted_labels = []
for i in range(len(predictions)):
  predicted_labels.append(int(predictions[i]['label'][-1]))

# compute metrics and print results
evaluator = evaluate.evaluator("text-classification")

metrics = ["accuracy", "f1", "precision", "recall"]
results = []

for m in metrics:
  metric = load_metric(m)
  if m == "accuracy":
    results.append(metric.compute(predictions=predicted_labels, references=ref_vars1))
  else:
    results.append(metric.compute(predictions=predicted_labels, references=ref_vars1, average="weighted"))

for i in range(len(metrics)):
  print("The " + metrics[i] + " is " + str(results[i]) + ".")

mark_pos = 0

for i in range(len(predicted_labels)):
  if predicted_labels[i] == 1:
    mark_pos += 1

print(mark_pos/1000)
```
